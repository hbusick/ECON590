```python
alpha = 0.3 # capitla share in production
delta = 0.1 # depreciation of capital 
beta = 0.9 # patience 
gamma = 2

Kss = ((1/ beta + delta - 1) / alpha) ** (1 / (alpha - 1))
print("Kss = ", Kss)
```

```python
import numpy as np

nK = 991

Kgrid = Kss * np.linspace(.01, 1.99, nK)
```
