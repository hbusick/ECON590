**How can we define a car class?**
The basic syntax for defining a new class in Python is the following: 

````python
class ClassName:
    # Class attribute (shared by all instances)
    class_attribute = "some value"

    def __init__(self, instance_attribute):
        # Instance attribute
        self.instance_attribute = instance_attribute

    def instance_method(self):
        return f"Hello, my  attribute is {self.instance_attribute}."
````

````python
a = ClassName("test")
````

````python
a.instance_attribute
````

What us happening here?  
1. ```class ClassName```:
   - This line defines a new class in Python called ```ClassName```.
   - A class is like a blueprint for creating objects that share common characteristics and behaviors.

2. ```class_attribute = "some value"```:
   - This is a class attribute, meaning it belongs to the class itself rather than any one object.
   - All instances (objects) of ```ClassName``` will share this attribute.
   - Changing ```class_attribute``` on the class level can affect how new instances see it.

3. ```def __init__(self, instance_attribute):```:  
   - ```__init__``` is a special method (constructor) called automatically when creating a new instance.
   - ```self``` represents the instance (object) being created.
   - ```instance_attribute``` is a parameter passed when constructing the class, e.g., ```obj = ClassName("my value")```.

4. ```self.instance_attribute = instance_attribute```:  
   - Inside ```__init__```, we assign the passed-in ```instance_attribute``` to the object’s ```self.instance_attribute```.
   - This becomes an instance attribute, meaning each object can store its own value.
   - Different objects can have different values for ```instance_attribute```.

5. ```def instance_method(self)```:  
   - A method that belongs to the class.
   - ```self``` refers to the instance on which the method is called.
   - Methods can access and manipulate the object’s data, such as ```self.instance_attribute```.

6. ```return f"Hello, my attribute is {self.instance_attribute}."```  
   - This line returns a formatted string referencing the specific object’s ```instance_attribute```.
   - If you create an object with `ClassName("Alice")`, calling `obj.instance_method()` would return `"Hello, I am Alice."`



````python
a = ClassName("test_a")
a.instance_method()

# we can also access the attributes directly:
print("a.instance_attribute:", a.instance_attribute)
print("a.class_attribute:", a.class_attribute)
````

````python
# we can also change the attributes directly:
a.instance_attribute = "new value"
print("a.instance_attribute:", a.instance_attribute)
````

````python
# we can crease more instances of the class:
b = ClassName("test_b")
print("b.instance_attribute:", b.instance_attribute)
print("b.class_attribute:", b.class_attribute)
````

### Inheritance
There can be child and parent classes, where each instance of the child class is a member of the parents class as well. The child class inherites attributes and methods from the parents class. For example, let's consider a Vehicle class. It should have the attributes ```year```, ```brand```, ```model```, ```num_wheels```, and ```engine```, as well as the methods ```start_engine``` and ```stop_engine```.

````python
class Vehicle:
    """A generic vehicle class."""

    def __init__(self, year, brand, model, num_wheels):
        self.year = year
        self.brand = brand
        self.model = model
        self.num_wheels = num_wheels
        self.engine = False  # Assume engine is off by default

    def start_engine(self):
        self.engine = True
        print(f"{self.brand} {self.model}'s engine is now running.")

    def stop_engine(self):
        self.engine = False
        print(f"{self.brand} {self.model}'s engine is now off.")
````

Now, say we want to define our ```Car``` class. Other than that cars (in this example) should always have 4 wheels, a lot of code would remain exactly as in the ```Vehicle``` class. By declaring the ```Car``` class as a child (or derived) class of the parent (or base) class ```Vehicle```, we can avoid dublicating code. We can also add a method that would only apply to cars, but not to all vehicles.

````python
class Car(Vehicle):
    """A car is a specific type of vehicle with 4 wheels."""

    def __init__(self, year, brand, model):
        # Call the parent class (Vehicle) constructor
        super().__init__(year, brand, model, num_wheels=4) #super means go to the parent class 
        #This calls the initialization of the parent class with a numwheel argument of 4

    # Car can have its own specialized methods or override parent methods
    def open_trunk(self):
        print(f"Opening the trunk of the {self.brand} {self.model}.")
````

### Dunder Methods
Dunder methods is short for double underscore methods. Sometimes we would like to be able to use operators like `+`, `-`, `*`, `==`, `<`, ... for objects from our class or we would like to print something more informative. To do this, we use so called dunder methods.


````python
class Vehicle:
    """A generic vehicle class."""

    def __init__(self, year, brand, model, num_wheels):
        self.year = year
        self.brand = brand
        self.model = model
        self.num_wheels = num_wheels
        self.engine = "off"  # Assume engine is off by default

    def start_engine(self):
        self.engine = "on"
        print(f"{self.brand} {self.model}'s engine is now running.")

    def stop_engine(self):
        self.engine = "off"
        print(f"{self.brand} {self.model}'s engine is now off.")

    def __str__(self):
        """
        __str__ defines what should be returned when you convert
        a Vehicle object to a string, for example when calling str(vehicle)
        or printing the object in a print statement.
        """
        return f"{self.year} {self.brand} {self.model} (Engine: {self.engine})"
    def __eq__(self, other):
        """
        __eq__ is called when using the == operator. This method
        defines the criteria for equality between two Car objects.
        """
        if not isinstance(other, Car):
            return False  # Different types can't be equal
        return (self.brand == other.brand and
                self.model == other.model and
                self.year == other.year)

    def __lt__(self, other):
        """
        __lt__ (less than) is called when using the < operator.
        Here we might define that a 'smaller' car is an older car
        (i.e., has an earlier 'year').
        """
        if not isinstance(other, Car):
            return NotImplemented
        
        return self.year < other.year
    
# now let's redefine the Car and Motorbike class
class Car(Vehicle):
    """A car is a specific type of vehicle with 4 wheels."""

    def __init__(self, year, brand, model):
        # Call the parent class (Vehicle) constructor
        super().__init__(year, brand, model, num_wheels=4)

    # Car can have its own specialized methods or override parent methods
    def open_trunk(self):
        print(f"Opening the trunk of the {self.brand} {self.model}.")

class Motorbike(Vehicle):
    """A motorbike is a type of vehicle with 2 wheels."""

    def __init__(self, year, brand, model):
        # Call Vehicle's constructor with num_wheels set to 2
        super().__init__(year, brand, model, num_wheels=2)

    def pop_wheelie(self):
        if self.engine == "on":
            print(f"{self.brand} {self.model} is popping a wheelie!")
        else:
            print(f"Cannot pop a wheelie with the engine off.")
        
````

### Encapsulation

#### What Is Encapsulation? 
- **Goal**: Hide internal data and implementations from the outside world (other parts of the program) and expose only what is necessary or safe to access.  
- **Example**: A `BankAccount` class may have a private or protected balance attribute and provide public methods like `deposit()` and `withdraw()` to interact with that balance.

We use **naming conventions** like a single underscore (`_balance`) for protected and a double underscore (`__balance`) for private attributes or methods. Protected or private attributes or methods should only be accessed or modified inside the class. Protected attributes or methods can still be accessed or modified directly, but indicate that we should not do it.

#### Why Do We Use It?

1. **Safety and Integrity**  
   - Prevent unintended changes to data. If an attribute is only updated through controlled methods, we can validate inputs and maintain data consistency.

2. **Simplicity**  
   - It’s easier to reason about code when you know exactly how data can be accessed or changed, reducing the risk of bugs and side effects.

3. **Maintainability**  
   - Lets you update internal implementation without breaking external code that relies on the public interface.

4. **Readability**  
   - By exposing only what’s necessary, you convey how the class is meant to be used. A well-encapsulated class shows clear valid actions (public methods) and hidden details (private or protected attributes).


**Example**: Write a `BankAccount` class, where the `balance` attribute is *private*. Users can not interact with it directlty but only through class methods.

````python
class BankAccount:
    """
    Represents a simple bank account with basic operations.
    Demonstrates how we can encapsulate data (balance) and control access
    through public methods (deposit, withdraw, and get_balance).
    """

    def __init__(self, account_holder, initial_balance=0):
        """
        Initializes the bank account with an account holder's name
        and an optional initial balance (default is 0).
        """
        # This is the name of the account holder (public attribute).
        self.account_holder = account_holder

        # The double underscore indicates that __balance is private and con not be accessed or modified directly from outside.
        self.__balance = initial_balance
        

    def deposit(self, amount):
        """
        Adds a positive amount to the account balance.
        If the amount is invalid (e.g., negative), it prints an error message.
        """
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit must be positive.")

    def withdraw(self, amount):
        """
        Withdraws a positive amount from the account if sufficient funds are available.
        Otherwise, it prints an error message indicating a denial.
        """
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Withdrawal denied: check your balance or the amount.")

    def get_balance(self):
        """
        Returns the current balance of the account.
        This method acts as a 'getter', controlling how balance is accessed.
        """
        return self.__balance
````

````python
myaccount = BankAccount("Econ590", 10)
# we can do this:
print("balance is ", myaccount.get_balance())  # 10
````

````python
# we can't do this:
print("balance is ", myaccount.__balance)  
````

````python
# we can only do this
myaccount.deposit(100)  # Deposited $100. New balance: $110
print("balance is ", myaccount.get_balance()) 
````

````python
# but not this (we could if balance would be protected (i.e. _ instead of __) instead of private)
myaccount.__balance += 100
````

````python
class Student:
    def __init__(self, name, major, gpa=0.0):
        self.name = name  # Public attribute
        self.major = major  # Public attribute
        self.__GPA = gpa  # Private attribute

    # Method to set the GPA
    def set_gpa(self, gpa):
        if 0.0 <= gpa <= 4.0:  # Validate GPA is within the correct range
            self.__GPA = gpa
        else:
            print("Error: GPA must be between 0.0 and 4.0.")

    # Method to get the GPA
    def get_gpa(self):
        return self.__GPA

    # Method to print information
    def print_info(self):
        print(f"Name: {self.name}")
        print(f"Major: {self.major}")
        print(f"GPA: {self.__GPA:.2f}")
````

````python
student1 = Student("Alice", "Economics", 3.5)
student1.print_info()  

student1.set_gpa(5.0)
student1.print_info()
````
