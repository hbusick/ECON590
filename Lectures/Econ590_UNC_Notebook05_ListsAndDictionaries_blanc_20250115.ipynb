### Lists
A **list** is an *ordered* and *mutable* collection that allows us to store elements of different data types and allows for *duplicate* elements.  

*Ordered* means that the order of the elements is preserved. You can think of this as books on a bookshelf: you can refer to the second book and the third book and so on. This is different to books in a big bag, where you know that the books are in there, but there is no obvious way to order them. 

*Mutable* means that elements can be added, removed, or changed. E.g. we can take book 4 from the bookshelf, take it out and replace it by another book. Or we can take book 4 from the shelf, and throw it away completely. In this case the book that was previously book 5 will now become book 4. Analogously we can insert a new book to the very left, becoming the new book 1. If we want, we can even take out book 4 and put something else in it's place. Like a little Mercedes toy car for example. Now the item number 4 is the toy mercedes, but all other items are books. A mutable object refers to a place in memory. Like refering to a specific bookshelf in our house. When a mutable object is modified, the changes occur in the same memory location without creating a new object.
This is why modifications to a mutable object via one reference will be visible to all other references pointing to the same object. If we have two mutable objects refering to the same bookshelf, if we change a book in that shelf, it will change for all objects pointing to that shelf.

*Duplicate* elements means that I can have the same element in multiple places of the list. To explain this, I could not think of a good bookshelf analogy, since lists would allow you to have the same physical book in multiple places on the shelf, not just different copies of the same book but literally the same book. As a counterexample you can maybe think of a regular language dictionary, where each word only appears once. More on dictionaries in python will come later.

```python
fruitIlike:list = ["Apple", "Pear", "Pineapple"]
fruitNOlike:list = ["Banana", "Coconut", "Grape"]

def doIlike(fruit:str) -> str:
    """
    Takes a fruit as an input and returns whether or not I like that fruit
    """
    if fruit in fruitIlike:
        return "I like that fruit"
    elif fruit in fruitNOlike:
        return "I do not like that fruit"
    else:
        return "I do not recognize that fruit"
```

#### IMPORTANT DIFFERENCE TO MATLAB: copies of lists and lists of mutable objects
So far we just looked at lists of strings and numbers. Strings and numbers are *immutable* objects. However, we can also have lists of mutable objects such as lists of lists. In this case it is important to know that for mutable objects the list stores references to the place of the objects in memory. So if we change those values, the list will change. Let's consider an example.

```python
a = "Econ"
b = 590
c = ["UNC <3"]

testlist = [a, b, c]
print("testlist = ", testlist)

# now, what happens if we change the value of a?
print("type(a) = ", type(a))
a = "Math"
```

This is because lists are mutable objects and point to places in memory.  
Similarly, if we just use `list1 = list2`, then they are pointing to the same bookshelf. So if I use one of the lists to change a book in the bookshelf, the book will also change in the other list. This can lead to dangerours mistakes. Let's consider an example. 

```python
testlist = ["USA", "Canada", "Mexico"]

testlist2 = testlist # this does not create a copy of testlist, it creates a reference to it

print("testlist = ", testlist)
print("testlist2 = ", testlist2)
```

```python
testlist2[0] = "Germany"
print("2 testlist = ", testlist)
print("2 testlist2 = ", testlist2)
print("BOTH LISTS CHANGED")
print("WHY? BECAUSE testlist2 IS A REFERENCE TO testlist, NOT A COPY OF IT")
print("id(testlist) = ", id(testlist))
print("id(testlist2) = ", id(testlist2))

```

Sometimes, we do not want that, we may want to be able to change elements in `testlist2` without changing elements in `testlist`. So we want those two lists to point to *two different places in memory*, which are initially filled with the same elements. We can do this by using `copy.deepcopy`. This will create a copy of the list in a different place in memory. This is like buying the same bookshelf and the same books and then putting it into a different room. Now I can change the books on one bookshelf without changing the books on the other. Let's try it.


```python
testlist = ["USA", "Canada", "Mexico"]


import copy # we need to import the copy module to use the deepcopy() function
testlist2 = copy.deepcopy(testlist) # this DOES create a copy of testlist

print("1 testlist = ", testlist)
print("1 testlist2 = ", testlist2)

testlist2[0] = "Germany"
print("2 testlist = ", testlist)
print("2 testlist2 = ", testlist2)
print("Now testlist remains unchanged")
```

### Tuples
A **tuple** is an *ordered*, *immutable* collection of elements in Python. Typles are similar to lists but their contents can not be changed after they are created. As for lists, we can put different types together into a tuple. We can create a tuple by using the round brackets `mytuple =  (a, b)` or by simply using a `,` only: `mytuple =  a, b`.

```python
mytuple = (1, 2, "Econ590")

print("mytuple = ", mytuple)
print("mytuple[0] = ", mytuple[0])
print("mytuple[1] = ", mytuple[1])
print("mytuple[2] = ", mytuple[2])
```

Tuples are immutable, so we can't change a specific entry of the tuple after creation.

```python

mytuple[0] = 3 # this will raise an error, because tuples are immutable
```

```python
# we can read out values from a tuple
a, b, c = mytuple
print("a = ", a)
print("b = ", b)
print("c = ", c)
```

We have already encountered tuples: if we define a function with multpile return values, the function actually returns a tuple.

```python
def compute_powers(number):
    return number**2, number**3, number**4

res = compute_powers(3)
```

```python

print("res = ", res)
print("type(res) = ", type(res))
print("res[0] = ", res[0])
print("res[1] = ", res[1])
print("res[2] = ", res[2])

# alternatively we can also do
a, b, c = compute_powers(3)
print("a = ", a)
print("b = ", b)
print("c = ", c)
```

### Dictionaries
A **dictionary** is a *mutable* collection of *key-value* pairs. Keys have to be immutable objects like numbers or strings and can't have duplicates. We can create and empy dictionary by using `{}` or fill it with values using `{key : value}` syntax. Keys and values do not need to be the same type and values are allowed to be mutable objects (like lists). Each key in a dictionary is unique and assigned to a value. 

```python
coursedict = {}

coursedict["Name"] = "Econ 590"
coursedict["Instructor"] = "Marlon"
coursedict["Location"] = "Gardner 211"
coursedict["Time"] = "TTh 3:30pm-4:45pm"
coursedict["NumStudents"] = 3
coursedict["Students"] = ["Alice", "Bob", "Charlie"]

print(coursedict)

# equivalently, we could have defined the dictionary as follows
coursedict = {"Name": "Econ 590", "Instructor": "Marlon", "Location": "Gardner 211", "Time": "TTh 3:30pm-4:45pm", "NumStudents": 3, "Students": ["Alice", "Bob", "Charlie"]}
print(coursedict)
```

```python
# a dictionary is useful because we can quickly access the value of a key
print(coursedict['Name'], "happens at", coursedict['Time'], "in", coursedict['Location'])

# we can change the value of a key
coursedict['Location'] = "Gardner 213"
print("Location is now" , coursedict['Location'])

# we can add a new key
coursedict['NumCredits'] = 3
print(coursedict)
```

```python
# items in a dictionary are key value pairs
for key, value in coursedict.items():
    print(key, ":", value)

# we can check if a key is in a dictionary
print("Name in coursedict = ", 'Name' in coursedict)
print("NumCredits in coursedict = ", 'NumCredits' in coursedict)
print("Grades in coursedict = ", 'Grades' in coursedict)
```

```python

# we can also iterate over the keys in a dictionary
for key in coursedict.keys():
    print('key = ', key)
    print('coursedict[key] = ', coursedict[key])
```

Values are allowed to be complicated objects, like nested lists or dictionaries. This can be useful. Assume we want to store the grades for each (fictional) student abd that their are grades for `"HW1"`, `"HW2"`, `"Midterm"`. For each student, I can store the grades in a dictionary. Then I can make another dictionary that maps each student name to their grade dictionary. And now I can store that grades dictionary into the course dictionary under (say) `"Grades"`. Let's see how this works.

```python
grades_Alice = {"HW1": 90, "HW2": 85, "Midterm": 80}
grades_Bob = {"HW1": 80, "HW2": 75, "Midterm": 70}
grades_Charlie = {"HW1": 70, "HW2": 65, "Midterm": 60}

grades = {"Alice": grades_Alice, "Bob": grades_Bob, "Charlie": grades_Charlie}

coursedict["Grades"] = grades
```

#### Hard exercise
Write code that extends the dictionaries at `coursedict['Grades']['Studentname']` with a key `'Average'` and a value equal to the average of the grades for `'HW1'`, `'HW2'`, and `'Midterm'`. 

```python
# hint
# loop through the students
for student in coursedict["Students"]: # iterate over the students in the course
    print("Grades for", student)

    # get the grades for the student
    grades = coursedict["Grades"][student]
    print("grades = ", grades)
    
    # read out the grades from the dictionary
    gradehw1 = grades["HW1"]
    gradehw2 = grades["HW2"]
    grademidterm = grades["Midterm"]

    print("HW1:", gradehw1)
    print("HW2:", gradehw2)
    print("Midterm:", grademidterm)    

    # compute the averge grade
    average_grade = (gradehw1 + gradehw2 + grademidterm) / 3

    # store the average grade in the dictionary
    coursedict["Grades"][student]["Average"] = average_grade

    print("Average Grade:", average_grade)
```

Commonly used methods for dictionaries include the following:

| **Method**        | **Description**                                   | **Example**                               |
|--------------------|---------------------------------------------------|-------------------------------------------|
| `get(key, default)`| Returns the value for the key, or `default`.      | `my_dict.get('key', 'default_value')`     |
| `keys()`           | Returns all keys as a view object.               | `my_dict.keys()`                          |
| `values()`         | Returns all values as a view object.             | `my_dict.values()`                        |
| `items()`          | Returns all key-value pairs as a view object.    | `my_dict.items()`                         |
| `update()`         | Updates the dictionary with key-value pairs.     | `my_dict.update({'new_key': 'value'})`    |
| `pop(key)`         | Removes and returns the value for the key.       | `my_dict.pop('key')`                      |
| `clear()`          | Removes all key-value pairs.                     | `my_dict.clear()`                         |


### Comparison
